## ALL THE COEFS.

```python
#LOAD DATA
data = list()
for i in range(1,9):
    data.append(np.loadtxt(f'./exports/col{i}.txt'))
```

```python
#CREATE PLOTS
fig = plt.figure(figsize=(20, 20))
for i in range(0,8):
    for j in range(0,8):
        ax = fig.add_subplot(9,9,i*9+j+1)
        ax.scatter(data[i], data[j],alpha=0.8, s=0.7)
#fig.suptitle('Distributions of Iris Features')
plt.show()
```
